# DS-2002 Data Project 1: ETL Data Pipeline

### **Project Overview**
This project builds a **dimensional data mart** to analyze **retail sales transactions**.  
It integrates **structured (MySQL), semi-structured (MongoDB), and file-based (CSV) data** and transforms them for **OLAP (Online Analytical Processing) analysis**.

### **Business Process Chosen: Retail Sales**
The data mart is designed for a **retail business**, tracking **sales transactions** with information on **customers, products, and purchase dates**.

### **ETL Pipeline**
1. **Extract** data from:
   - MySQL (`AdventureWorks` retail database)
   - MongoDB (`customer_data` NoSQL database)
   - CSV file (`products.csv` from local file system)
2. **Transform** the data (cleaning, renaming, standardizing).
3. **Load** the processed data into `AdventureWorks_DW` (data warehouse).

### **OLTP vs. OLAP**
- **OLTP (Transactional Data)** → Stored in `AdventureWorks` MySQL.
- **OLAP (Aggregated Data for Analysis)** → Processed & loaded into `AdventureWorks_DW`.

### Documentation
- **Extract**: Data is sourced from MySQL (AdventureWorks), MongoDB (customer_data), and a CSV file (products.csv).
- **Transform**: Data is cleaned (removing nulls, standardizing column names) and mapped to the dimensional schema.
- **Load**: Data is loaded into `AdventureWorks_DW` with tables `dim_customers`, `dim_products`, `dim_date`, and `fact_sales`.
- **Deployment**: MySQL and MongoDB are hosted locally. All code is in this notebook, and data files are included in the GitHub repo.
- **Analysis**: SQL queries provide insights into customer spending and product category performance over time.

#### Define Functions for Getting Data From and Setting Data Into Databases

```python
# helper function from lab 3 and 4
def get_sql_dataframe(sql_query, **args):
    '''Create a connection to the MySQL source database and return a Pandas DataFrame.'''
    conn_str = f"mysql+pymysql://{args['uid']}:{args['pwd']}@{args['hostname']}:{args['port']}/{args['src_dbname']}"
    sqlEngine = create_engine(conn_str, pool_recycle=3600)
    connection = sqlEngine.connect()
    
    # Execute Query and Fetch Data
    dframe = pd.read_sql(text(sql_query), connection)
    connection.close()
    
    return dframe


def set_dataframe(df, table_name, pk_column, db_operation, **args):
    '''Insert or update data in the MySQL destination database using Pandas DataFrame.'''
    conn_str = f"mysql+pymysql://{args['uid']}:{args['pwd']}@{args['hostname']}:{args['port']}/{args['dst_dbname']}"
    sqlEngine = create_engine(conn_str, pool_recycle=3600)
    connection = sqlEngine.connect()
    
    if db_operation == "insert":
        df.to_sql(table_name, con=connection, index=False, if_exists='replace')
        connection.execute(text(f"ALTER TABLE {table_name} ADD PRIMARY KEY ({pk_column});"))
            
    elif db_operation == "update":
        df.to_sql(table_name, con=connection, index=False, if_exists='append')
    
    connection.close()


def get_mongo_client(**args):
    '''Validate MongoDB connection and return a client instance.'''
    if args["cluster_location"] not in ['atlas', 'local']:
        raise Exception("You must specify either 'atlas' or 'local' for the cluster_location parameter.")
    
    else:
        if args["cluster_location"] == "atlas":
            connect_str = f"mongodb+srv://{args['user_name']}:{args['password']}@"
            connect_str += f"{args['cluster_name']}.{args['cluster_subnet']}.mongodb.net"
            client = MongoClient(connect_str)
            
        elif args["cluster_location"] == "local":
            client = MongoClient("mongodb://localhost:27017/")
        
    return client


def get_mongo_dataframe(mongo_client, db_name, collection, query={}):
    '''Query MongoDB, retrieve documents, and return a Pandas DataFrame.'''
    db = mongo_client[db_name]
    dframe = pd.DataFrame(list(db[collection].find(query)))
    
    # Drop MongoDB's default `_id` column
    if "_id" in dframe.columns:
        dframe.drop(["_id"], axis=1, inplace=True)
    return dframe


def set_mongo_collections(mongo_client, db_name, data_directory, json_files):
    '''Load JSON data from a directory and insert into MongoDB collections.'''
    db = mongo_client[db_name]
    
    for file in json_files:
        db.drop_collection(file)  # Drop existing collection to avoid duplicates
        json_file = os.path.join(data_directory, json_files[file])
        with open(json_file, 'r') as openfile:
            json_object = json.load(openfile)
            file = db[file]
            file.insert_many(json_object)
        
    #mongo_client.close()
    
```

```python
# SQL query to calculate total sales per customer and product, with correct reference to ProductID
sql_sales_query = """
SELECT c.customer_key, c.CustomerID, p.product_key, p.Name AS product_name, 
       COUNT(*) AS total_orders, SUM(fs.total_sales) AS total_revenue
FROM adventureworks_dw.fact_sales AS fs
INNER JOIN adventureworks_dw.dim_customer AS c ON fs.customer_key = c.customer_key
INNER JOIN adventureworks_dw.dim_product AS p ON fs.ProductID = p.product_id  -- Corrected the join
GROUP BY c.customer_key, c.CustomerID, p.product_key, p.Name
ORDER BY total_revenue DESC
LIMIT 10;
"""

# Fetch the results and display the first 5 rows
df_sales_summary = get_sql_dataframe(sql_sales_query, **mysql_args)
display(df_sales_summary)
```

```python
# SQL query to calculate total sales, average order quantity per customer and product
sql_sales_query = """
SELECT 
    c.customer_key, 
    c.CustomerID, 
    p.product_key, 
    p.Name AS product_name, 
    COUNT(*) AS total_orders, 
    SUM(fs.TotalDue) AS total_revenue,
    AVG(fs.OrderQty) AS avg_order_qty
FROM 
    adventureworks_dw.fact_sales AS fs
INNER JOIN 
    adventureworks_dw.dim_customer AS c ON fs.customer_key = c.customer_key
INNER JOIN 
    adventureworks_dw.dim_product AS p ON fs.ProductID = p.product_id  -- Adjusted join to use ProductID
GROUP BY 
    c.customer_key, c.CustomerID, p.product_key, p.Name
ORDER BY 
    total_revenue DESC
LIMIT 10;
"""

# Fetch the results and display the first 5 rows
df_sales_summary = get_sql_dataframe(sql_sales_query, **mysql_args)
display(df_sales_summary)
```

- **Gold**: Aggregated monthly product sales

### Tools Used
- Apache Spark Structured Streaming (via PySpark)
- CSV + JSON files for batch and streaming data

## Project Initialization & Spark Setup

This project implements a Lakehouse architecture in Spark for sales data, following the Bronze → Silver → Gold pipeline pattern.

- We start by setting environment variables (`JAVA_HOME`, `SPARK_HOME`) and initializing a `SparkSession` with schema inference enabled.
- Directory paths are established for data, streaming inputs, and warehouse storage.
- JSON files are split into multiple batches to simulate real-time ingestion.

Key variables initialized:
- `base_dir`, `data_dir`, `stream_dir`, `orders_stream_dir`
- `warehouse_dir`, `dest_database`
- Output paths for `bronze`, `silver`, and `gold` layers.

#### Silver Layer: Cleaning and Enrichment

We read the raw Bronze data and join it with dimension tables to enrich and standardize it.

Transformations:
- Dropped duplicate date keys (`order_date_key_y`) and renamed `order_date_key_x` for clarity.
- Joined with:
  - `dim_customers` using `customer_key`
  - `dim_products` using `ProductID`
  - `dim_date` using `order_date_key`
- Added `order_full_date` from `dim_date`
- Output includes `TotalDue`, keys, metadata, and enriched date info.

The cleaned, enriched fact data is streamed to the **Silver** layer at:


```python
from pyspark.sql.functions import col

# Load Bronze data as a static DataFrame
df_sales_bronze_loaded = spark.read.format("parquet").load(sales_output_bronze)

# Drop and rename date keys to simplify join
bronze_clean = df_sales_bronze_loaded \
    .drop("order_date_key_y") \
    .withColumnRenamed("order_date_key_x", "order_date_key")

# Prepare dim_date for join
df_order_date = df_dim_date \
    .withColumnRenamed("date_key", "order_date_key") \
    .withColumnRenamed("full_date", "order_full_date")

# Join Bronze + Dimension tables to create Silver layer
df_sales_silver = bronze_clean \
    .join(df_dim_customer, on="customer_key", how="inner") \
    .join(df_dim_product, bronze_clean["ProductID"] == df_dim_product["product_id"], how="inner") \
    .join(df_order_date, on="order_date_key", how="left") \
    .select(
        col("sales_id").alias("sales_order_id"),
        col("customer_key"),
        col("product_key"),
        col("TotalDue"),
        col("order_date_key"),
        col("order_full_date"),
        col("receipt_time"),
        col("source_file")
    )

# Save Silver output as Parquet (batch write)
df_sales_silver.write \
    .format("parquet") \
    .mode("overwrite") \
    .option("compression", "snappy") \
    .save(sales_output_silver)

print(" Silver layer successfully written to:", sales_output_silver)
```

#### Gold Layer: Monthly Customer Spending by Country

In this final aggregation step, we calculated the total amount each customer spent per month, grouped by their country. This provides a geographic and temporal view of customer value — helpful for regional sales strategies or customer segmentation.

### Transformations:
In the final stage, we aggregate sales by:
- `CustomerID`
- `Country_Region`
- `month_name`


```python
from pyspark.sql.functions import month, date_format, sum as _sum, desc

# Load Silver layer
df_sales_silver = spark.read.format("parquet").load(sales_output_silver)

# Load updated dimension tables
df_dim_date = spark.sql(f"SELECT * FROM {dest_database}.dim_date")
df_dim_product = spark.sql(f"SELECT * FROM {dest_database}.dim_products")
df_dim_customer = spark.sql(f"SELECT * FROM {dest_database}.dim_customers")

# Add month metadata
df_dim_date = df_dim_date \
    .withColumn("month_of_year", month("full_date")) \
    .withColumn("month_name", date_format("full_date", "MMMM"))

# Join and aggregate by customer country + month
df_gold_custom = df_sales_silver \
    .join(df_dim_product, on="product_key", how="inner") \
    .join(df_dim_customer, on="customer_key", how="inner") \
    .join(df_dim_date, df_sales_silver.order_date_key == df_dim_date.date_key, "inner") \
    .groupBy("month_name", "month_of_year", "CustomerID", "Country_Region") \
    .agg(_sum("TotalDue").alias("customer_total_spent")) \
    .orderBy("month_of_year", desc("customer_total_spent"))

# Define table name and path
custom_gold_table = f"{dest_database}.gold_customer_spending_by_country"
gold_path = os.path.join(warehouse_dir, f"{dest_database}.db", "gold_customer_spending_by_country")

# Clean up existing path before writing
if os.path.exists(gold_path):
    shutil.rmtree(gold_path)
    print(f"Deleted existing path: {gold_path}")

# Save to Gold managed table
df_gold_custom.write.mode("overwrite").saveAsTable(custom_gold_table)

# Display result
print(" Final: Customer Spending by Country and Month")
spark.sql(f"SELECT * FROM {custom_gold_table} ORDER BY month_of_year DESC, customer_total_spent DESC").show(15, truncate=False)
```
